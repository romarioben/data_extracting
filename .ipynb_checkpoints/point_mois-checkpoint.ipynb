```python
import pandas as pd
import numpy as np
from datetime import date
```

```python
df = pd.read_excel("POINT_JUILLET_2024.xlsx")
df
```

```python
list_index = []
for index, line in df.iterrows():
    if line["NOMS"] is np.NaN:
        list_index.append(index)
df = df.drop(list_index, axis=0)
print(list_index)
df 
```

```python
Nat = type(df["DATE"][0])
print(type(df["DATE"][6]) == Nat)
print(Nat)
```

```python
#### Remplir les dates non Ã©crites 
for index, line in df.iterrows():
    if type(line["DATE"]) == type(df["DATE"][0]):
        datee = line["DATE"]
        #print(datee)
    else:
        #line["DATE"] = datee
        df.at[index, "DATE"] = datee
df
```

```python
dict_dates = {}
for index, line in df.iterrows():
    datee = line["DATE"]
    dict_dates[datee] = dict_dates.get(datee,{})
    examens = line["EXAMENS"]
    examens = examens.split()
    print(examens)
    for examen in examens :
        #print(examen)
        dict_dates[datee][examen] = dict_dates[datee].get(examen, 0) + 1
    print(datee, dict_dates[datee])
print(dict_dates)
```

```python
dict_par_dates = {"DATES" : [], "LIBELLÃ‰": [], "QTE": []}
dict_complet = {}
for dato, para in dict_dates.items():
    for examen, valeur in para.items():
        dict_par_dates["DATES"].append(dato.strftime("%d-%m-%Y"))
        dict_par_dates["LIBELLÃ‰"].append(examen)
        dict_par_dates["QTE"].append(valeur)
        dict_complet[examen] = dict_complet.get(examen, 0) + valeur 
df1 = pd.DataFrame(dict_par_dates)

print(dict_complet)
dict_complet_frame = {"ParamÃ¨tres":[], "Valeur":[]}
for cle, valeur in dict_complet.items():
    dict_complet_frame["ParamÃ¨tres"].append(cle)
    dict_complet_frame["Valeur"].append(valeur)

#### Mettre par ordre alphabÃ©tique les paramÃ¨tres 
dict_complet_alph = {"ParamÃ¨tres":[], "Valeur":[]}
alpha_para = list(dict_complet.keys())
alpha_para.sort()
for para in alpha_para:
    dict_complet_alph["ParamÃ¨tres"].append(para)
    dict_complet_alph["Valeur"].append(dict_complet[para])
df2 = pd.DataFrame(dict_complet_alph)
df2
```

```python
df1.to_excel("point_journalier_assurance_juin_2024.xlsx", index=False)
df2.to_excel("point_complet_assurance_juin_2024.xlsx", index=False)
```
